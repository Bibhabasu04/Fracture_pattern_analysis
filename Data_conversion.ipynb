```python
import numpy as np
import pandas as pd
```

```python
file = np.array(pd.read_csv("D:/Projects/Geology Project/FRACTURE 15.2.16.csv"))
file = file[:,1:4]
n, p = file.shape
```

```python
strike, dip, direction = file[:,0], file[:,1], file[:,2]
```

```python
empty = []
D = ['N', 'E', 'S','W']
for i in range(n):
    if direction[i] not in D:
        empty.append(i)
        direction[i] = 0    
    for j in range(4):
        if direction[i] == D[j]:
            direction[i] = 90*j
empty
```

```python
new_dir = np.zeros(n)
for i in range(n):
    s1 = strike[i] - 90 + ((strike[i] - 90) < 0)*360
    s2 = strike[i] + 90 - ((strike[i] + 90) > 360)*360
    if abs(direction[i] - s1) < abs(direction[i] - s2):
        new_dir[i] = s1
    else:
        new_dir[i] = s2
```

```python
file =  open('fractures_dip_direction.txt', 'w')
for c1, c2 in zip(new_dir, dip):
    file.write(f"{c1}\t{c2}\n")
```
